Для моделирования используется схема БИС (Бикомпактная Интерполяционная Схема), представляющая собой модификацию схемы CIP (Cubic Interpolated Profile). Здесь $u,v -$ массив для интерполируемой величины и её производной, $i -$ номер узла, в котором производится интерполяция, $l>0 -$ скорость звука в среде,
$c$ принимает значение 1 или -1 в зависимости от того, в какую сторону двигается возмущение (1, если вправо, -1, если влево).

```python
def CIR(u,v,i,l,c):
    a=(u[1,i]-u[1,i-c])/h
    b=u[1,i] 
    arg=(-1)*l*tau
    u[0,i]=a*arg+b
    v[0,i]=a
```

```python
%%cython
import numpy as np
cimport numpy as np
DTYPE = np.float 
ctypedef np.float_t DTYPE_t

def cip_c(float u, float v, float u1, float v1, float la, float h, float dt):
    "la is wavespeed"
    cdef float a,b,c,d,delta0,uu,vv
    delta0 = u - u1
    a = (v + v1) / h / h - 2.0 * delta0 / h / h / h
    b = (2.0 * v + v1) / h - 3.0 * delta0 / h / h
    c = v
    d = u
    x = -la * dt
    uu = a * x * x * x + b * x * x + c * x + d
    vv = 3.0 * a * x * x + 2.0 * b * x + c
    return uu,vv

def CIRC(float a, float c, float b, float d, float Vp,float h,float dt):
    "a = u[1,i], b= u[1,i-c], c= v[1,i], d=v[1,i-c], Vp is wavespeed, h, tau are steps"
    cdef float i,arg,u,v
    i=(a-b)/h
    arg=(-1)*Vp*dt
    u=i*arg+a
    v=i
    return u,v

def CIP2RC(float a, float b, float c, float d, float Vp,float h,float dt):
    "a = u[1,i], b= u[1,i-c], c= v[1,i], d=v[1,i-c], Vp is wavespeed, h, tau are steps"
    cdef float i,j,k,arg,u,v
    i=(-1)*(a-b)/h/h+c/h
    j=c
    k=a
    arg=(-1)*Vp*dt
    u=i*arg*arg+j*arg+k
    v=2*i*arg+j
    return u,v

def CIP2LC(float a, float b, float c, float d, float Vp,float h,float dt):
    "a = u[1,i], b= u[1,i-c], c= v[1,i], d=v[1,i-c], Vp is wavespeed, h, tau are steps"
    cdef float i,j,k,arg,u,v
    i=(a-b)/h/h-d/h
    j=2*(a-b)/h-d
    k=a
    arg=(-1)*Vp*dt
    u=i*arg*arg+j*arg+k
    v=2*i*arg+j
    return u,v

def CIPC(float u, float u1, float v, float v1, float la, float h, float dt):
    "la is wavespeed"
    cdef float a,b,c,d,delta0,uu,vv
    delta0 = u - u1
    a = (v + v1) / h / h - 2.0 * delta0 / h / h / h
    b = (2.0 * v + v1) / h - 3.0 * delta0 / h / h
    c = v
    d = u
    x = -la * dt
    uu = a * x * x * x + b * x * x + c * x + d
    vv = 3.0 * a * x * x + 2.0 * b * x + c
    return uu,vv

def R_C(float p, float u, float zl, float zr):
    cdef float w1,w2
    if zl+zr==0:
        return 0, 0
    w1 = (- p + zr * u) / (zl + zr)
    w2 = (p + zl * u) / (zl + zr)
    return w1, w2

def RC(float w1, float w2, float zl, float zr):
    cdef float u,p
    p = (- w1 * zl + w2 * zr)
    u = (w1 + w2)
    return p, u

def R2D_C():
    return

def R2DC():
    return

def BIS1C(float u, float v, float u1, float v1, float Vp,float h,float dt):
    cdef float uu,vv
    uu,vv=CIPC(u,u1,v,v1,Vp,h,dt)
    if uu<min(u,u1) or uu>max(u,u1):
        uu,vv=CIP2LC(u,u1,v,v1,Vp,h,dt)
    if uu<min(u,u1) or uu>max(u,u1):
        uu,vv=CIP2RC(u,u1,v,v1,Vp,h,dt)
    if uu<min(u,u1) or uu>max(u,u1):
        uu,vv=CIRC(u,v,u1,v1,Vp,h,dt)
    return uu,vv

def BIS2C(float u, float v, float u1, float v1, float Vp,float h,float dt):
    cdef float vz, aa,bb,x0, uu, vv
    vz=(u-u1)/2 
    aa=(2*v+v1)/h-3*(u-u1)/h/h
    bb=(v+v1)/h/h-2*(u-u1)/h/h/h
    x0=aa/(2*bb+1e-06) #координата вершины параболы для обнаружения возможного экстремума на рассматриваемом интервале

    if   v*vz>=0   and   v1*v>=0: #условие (12) из статьи
        if x0>-h and x0<0 and (3*aa*x0*x0+2*bb*x0+v)*vz<0: #условие (13) из статьи
            if min(v,v1)<vz and vz < max(v,v1):
                uu,vv=CIRC(u,v,u1,v1,Vp,h,dt)
            elif abs(v1-vz)<abs(v-vz):     #условие (16) из статьи
                uu,vv=CIP2LC(u,u1,v,v1,Vp,h,dt)
            else:
                uu,vv=CIP2RC(u,u1,v,v1,Vp,h,dt)
        else:
            uu,vv=CIPC(u,u1,v,v1,Vp,h,dt)
                #CIP(u[1,i],u[1,i-c],v[1,i],v[1,i-c],l[i],h,tau)
    elif v*v1>=0 and vz*v<0: # условие (18) из статьи 
        uu,vv=CIRC(u,v,u1,v1,Vp,h,dt)
    else:
        if abs(v1-vz)<=abs(v-vz):    
            uu,vv=CIP2LC(u,u1,v,v1,Vp,h,dt)

        else:   
            uu,vv=CIP2RC(u,u1,v,v1,Vp,h,dt)
    #uu,vv=CIPC(u,v,u1,v1,Vp,h,dt)
    return uu,vv
```

```python
def string_schema(p1, u1, dp1, du1, p2, u2, dp2, du2, z, cp, h, dt, schema):
    "осуществляет обсчёт одномерного уравнения акустики с помощью схемы schema"
    for i in range(1, int(N)-1):
        zl = z[i]
        zr = z[i-1]
        w01, w02 = R_C(p1[i], u1[i], zl, zr)
        w_11, w_12 = R_C(p1[i-1], u1[i-1], zl, zr)
        w11, w12 = R_C(p1[i+1], u1[i+1], zl, zr)
            
        dw01, dw02 = R_C(dp1[i], du1[i], zl, zr)
        dw_11, dw_12 = R_C(dp1[i-1], du1[i-1], zl, zr)
        dw11, dw12 = R_C(dp1[i+1], du1[i+1], zl, zr)
        if schema=="BIS1":
            w01, dw01 = BIS1C(w01, dw01, w11, dw11, cp[i], h, dt)
            w02, dw02 = BIS1C(w02, dw02, w_12, dw_12, cp[i-1],h, dt)
        if schema=="CIP":
            w01, dw01 = cip_c(w01, dw01, w11, dw11, cp[i], h, dt)
            w02, dw02 = cip_c(w02, dw02, w_12, dw_12, cp[i-1],h, dt)
        if schema=="BIS2":
            w01, dw01 = BIS2C(w01, dw01, w11, dw11, cp[i], h, dt)
            w02, dw02 = BIS2C(w02, dw02, w_12, dw_12, cp[i-1],h, dt)
        if schema=="CIR":
            w01, dw01 = CIRC(w01, dw01, w11, dw11, cp[i], h, dt)
            w02, dw02 = CIRC(w02, dw02, w_12, dw_12, cp[i-1],h, dt)               
        p2[i], u2[i] = RC(w01, w02, zl, zr)
        dp2[i], du2[i] = RC(dw01, dw02, zl, zr)
    return #p2, u2, dp2, du2 
```

```python
def cip1(p, dp, u, du, nt, dt, dx, z, cp, schema):
    pn = np.zeros((nt,len(p)))
    pn[:,:] = p.copy()

    p1 = p.copy()    
    u1 = u.copy()
    du1 = du.copy()
    dp1 = dp.copy()
    
    for j in range(1, nt):
        u2 = u1.copy()
        p2 = p1.copy()
        du2 = du1.copy()
        dp2 = dp1.copy()
        string_schema(p1, u1, dp1, du1, p2, u2, dp2, du2, z, cp, h, dt, schema)
        p1 = p2.copy()
        u1 = u2.copy()
        dp1 = dp2.copy()
        du1 = du2.copy()       
        pn[j] = p1.copy()
    return pn
```
